```python
from bs4 import BeautifulSoup
import requests
Contact = []
for i in range(0,20):
    url = 'https://www.bizbaya.com/sme-india/rajasthan/jaipur?page='
    url = url + str(i)

    source = requests.get(url).text
    soup = BeautifulSoup(source,'lxml')
    table=soup.find('div',{'class':'view-content'})
    
    for tr in table.find_all('tr'):
        try:
            Mobile = tr.find("p",{"class":"views-field views-field-mobile"})
            Mobile = Mobile.span.text
        except Exception as e:
            Mobile = None
        Contact.append(Mobile)
print(Contact)
```

```python
from bs4 import BeautifulSoup
import requests
import csv

csv_file = open('BizbayaData.csv', 'w')
csv_writer= csv.writer(csv_file)
csv_writer.writerow(['Company', 'Director', 'Mobile', 'Telephone', 'Fax', 'Email', 'Address', 'Pin'])


for i in range(0,20):
    url = 'https://www.bizbaya.com/sme-india/rajasthan/jaipur?page='
    url = url + str(i)
    
    source = requests.get(url).text
    soup = BeautifulSoup(source,'lxml')
    table = soup.find("div",{"class":"view-content"})

   
    for tr in table.find_all('tr'):
        try:
            Cname = tr.h3.a.text
        except Exception as e:
            Cname = None
        print(Cname)
    
        try:
            Dname = tr.find("p",{"class","views-field views-field-directors"})
            Dname = Dname.span.text
        except Exception as e:
            Dname = None
        print(Dname)
    
        try:
            Mobile = tr.find("p",{"class":"views-field views-field-mobile"})
            Mobile = Mobile.span.text
        except Exception as e:
            Mobile = None
        print(Mobile)
    
        try:
            Telephone = tr.find("p",{"class":"views-field views-field-telephone"})
            Telephone = Telephone.span.text
        except Exception as e:
            Telephone = None
        print(Telephone)
            
        try:
            Fax = tr.find("p",{"class":"views-field views-field-fax"})
            Fax = Fax.span.text
        except Exception as e:
            Fax = None
        print(Fax)
    
        try:
            Email = tr.find("p",{"class":"views-field views-field-email"})
            Email = Email.span.text
        except Exception as e:
            Email = None
        print(Email)
    
        try:
            Address = tr.find("p",{"class","views-field views-field-address"})
            Address = Address.span.text
        except Exception as e:
            Address = None
        print(Address)
    
        try:
            Pin = tr.find("p",{"class":"views-field views-field-pin"})
            Pin = Pin.span.text
        except Exception as e:
            Pin = None
        print(Pin)
        
    csv_writer.writerow([Cname,Dname,Mobile,Telephone,Fax,Email,Address,Pin])

csv_file.close()
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


Company = []
Director = []
MobileNo = []
Tph = []
F = []
Mail = []
A = []
P = []

for i in range(0,20):
    url = 'https://www.bizbaya.com/sme-india/rajasthan/jaipur?page='
    url = url + str(i)
    
    source = requests.get(url).text
    soup = BeautifulSoup(source,'lxml')
    table = soup.find("div",{"class":"view-content"})
    
    for tr in table.findAll('tr'):
        try:
            Cname = tr.h3.a.text
        except Exception as e:
            Cname = None
        Company.append(Cname)
        #print(Cname)
    
        try:
            Dname = tr.find("p",{"class":"views-field views-field-directors"})
            Dname = Dname.span.text
        except Exception as e:
            Dname = None
        Director.append(Dname)
        #print(Dname)
    
        try:
            Mobile = tr.find("p",{"class":"views-field views-field-mobile"})
            Mobile = Mobile.span.text
        except Exception as e:
            Mobile = None
        MobileNo.append(Mobile)
        #print(Mobile)
    
        try:
            Telephone = tr.find("p",{"class":"views-field views-field-telephone"})
            Telephone = Telephone.span.text
        except Exception as e:
            Telephone = None
        Tph.append(Telephone)
        #print(Telephone)
            
        try:
            Fax = tr.find("p",{"class":"views-field views-field-fax"})
            Fax = Fax.span.text
        except Exception as e:
            Fax = None
        F.append(Fax)
        #print(Fax)
    
        try:
            Email = tr.find("p",{"class":"views-field views-field-email"})
            Email = Email.span.text
        except Exception as e:
            Email = None
        Mail.append(Email)
        #print(Email)
    
        try:
            Address = tr.find("p",{"class","views-field views-field-address"})
            Address = Address.span.text
        except Exception as e:
            Address = None
        A.append(Address)
        #print(Address)
    
        try:
            Pin = tr.find("p",{"class":"views-field views-field-pin"})
            Pin = Pin.span.text
        except Exception as e:
            Pin = None
        P.append(Pin)
        #print(Pin)

print(Company)
print(Director)
print(MobileNo)
print(Tph)
print(F)
print(Mail)
print(A)
print(P)
```

```python
import pandas as pd
df={'Comapny':Company,'Director':Director,'Mobile':MobileNo,'Telephone':Tph,'Fax':F,'Email':Mail,'Address':A,'Pin':P}
lst=pd.DataFrame(df)
lst.to_excel('new.xlsx')
```
